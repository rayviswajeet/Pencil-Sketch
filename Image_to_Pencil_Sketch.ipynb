#**Image to Pencil Sketch**

```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img = cv2.imread("/content/drive/MyDrive/Car.png")
```

```python
cv2_imshow(img)
```

#Converting Original Image into Grayscale

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

#Transformation of the greyscale image into negative using bitwise_not operator

```python
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

#Pencil Sketch from Original Image

```python
def dodgeV2(x, y):
  return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dodgeV2(img_gray, img_smoothing)
cv2_imshow(final_img)
```

#Original Vs Sketch Image

```python
cv2_imshow(img)
cv2_imshow(final_img)
```
